```python
import pandas as pd
import matplotlib.pyplot as plt
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
```

```python
# STATENT Datei 2021 herunterladen
url = urlopen("https://dam-api.bfs.admin.ch/hub/api/dam/assets/27245297/master")

# Entzippen und lesen der STATENT Daten
zipfile = ZipFile(BytesIO(url.read()))
dfSTATENTSource = pd.read_csv(zipfile.open('ag-b-00.03-22-STATENT2021/STATENT_GMDE_2021.csv'),
                         usecols=['ERHJAHR', 'GDENR', 'B08VZAT'],
                         sep=';')

dfSTATENTSource.head()
```

# Neuer Abschnitt

```python
# Kopfzeilen umbenennen
dfSTATENTSource = dfSTATENTSource.rename(columns={"ERHJAHR": "Jahr",
                              "GDENR": "BFS_Nummer",
                              "B08VZAT": "Vollzeitaeq"})
dfSTATENTSource.head()
```

```python
# Als CSV-Datei exportieren
dfSTATENTSource.to_csv('./sample_data/output_statent_2021.csv',sep=';', encoding="utf-8-sig")
```
